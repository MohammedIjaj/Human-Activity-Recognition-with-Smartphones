```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize the distribution of a few features
sns.histplot(train_data['tBodyAcc-mean()-X'], kde=True)
plt.show()

# Visualize the target variable distribution
sns.countplot(x='Activity', data=train_data)
plt.show()
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

# Select top 10 features
X = train_data.drop(['Activity', 'subject'], axis=1)
y = train_data['Activity']
selector = SelectKBest(f_classif, k=10)
X_new = selector.fit_transform(X, y)

selected_features = X.columns[selector.get_support(indices=True)]
print("Selected features:", selected_features)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X_new, y, test_size=0.2, random_state=42)

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_val)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
classification_report(y_val, y_pred)
confusion_matrix(y_val, y_pred)
```

```python
X_test = test_data[selected_features]
test_predictions = model.predict(X_test)

test_data['Predicted_Activity'] = test_predictions
test_data[['subject', 'Predicted_Activity']].head()
```
